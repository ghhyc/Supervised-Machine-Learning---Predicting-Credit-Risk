```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
test_df.head(10)
```

```python
# Convert categorical data to numeric and separate target feature for training data
y_train= train_df["loan_status"]
x_train= train_df.drop(columns=["loan_status"])
train_data = pd.get_dummies(train_df)  
```

```python
train_data.head(10)
```

```python
# Convert categorical data to numeric and separate target feature for testing data
y_test= test_df["loan_status"]
x_test= test_df.drop(columns=["loan_status"])
test_data = pd.get_dummies(test_df)  
test_data.head(10)
```

```python
# add missing dummy variables to testing set
dummies_to_add = set(train_data.columns) - set(test_data.columns)
for column_name in dummies_to_add:
    test_data[column_name] = 0
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score
x_train = train_data.drop(columns=['loan_status_high_risk', 'loan_status_low_risk'])
y_train = train_data['loan_status_high_risk']
regre =  LogisticRegression(solver='lbfgs', max_iter=10000).fit(x_train,y_train)
```

```python
x_test = test_data.drop(columns=['loan_status_high_risk', 'loan_status_low_risk'])
y_test = test_data['loan_status_high_risk']
```

```python
print(regre.score(x_test, y_test))
```

```python
# Train a Random Forest Classifier model and print the model score
clf = RandomForestClassifier(random_state=1).fit(x_train,y_train)
print(f'Training Score: {clf.score(x_train, y_train)}')
print(f'Testing Score: {clf.score(x_test, y_test)}')
```

```python
# Scale the data
scaler = StandardScaler().fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
 
```

```python
# Train the Logistic Regression model on the scaled data and print the model score
regre =  LogisticRegression(solver='lbfgs', max_iter=10000).fit(x_train_scaled,y_train)
print(f'Training Score: {regre.score(x_train_scaled, y_train)}')
print(f'Testing Score: {regre.score(x_test_scaled, y_test)}')
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
clf = RandomForestClassifier(random_state=1).fit(x_train_scaled,y_train)
print(f'Training Score: {clf.score(x_train_scaled, y_train)}')
print(f'Testing Score: {clf.score(x_test_scaled, y_test)}')
```
